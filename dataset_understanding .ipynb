```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
from collections import Counter
import nltk
from nltk.corpus import stopwords
from nltk import word_tokenize, pos_tag

# Download required NLTK data
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')

# Set of English stopwords
stop_words = set(stopwords.words('english'))

# Load dataset
df = pd.read_csv("/content/news_sentiment_analysis.csv")

# Drop NaNs in 'Description'
df['Description'] = df['Description'].astype(str)

# Add text length feature
df['TextLength'] = df['Description'].apply(len)

# 1. Sentiment Class Distribution
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Sentiment', palette='Set2')
plt.title("Sentiment Class Distribution")
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.show()

# 2. Text Length Distribution
plt.figure(figsize=(8,4))
sns.histplot(df['TextLength'], bins=40, color='skyblue')
plt.title("Distribution of Text Lengths")
plt.xlabel("Text Length")
plt.ylabel("Frequency")
plt.show()

# 3. Text Length vs Sentiment
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='Sentiment', y='TextLength', palette='Set3')
plt.title("Text Lengths by Sentiment")
plt.show()

# 4. Most Common Words per Sentiment
def get_top_words(sentiment, n=10):
    text = " ".join(df[df['Sentiment'] == sentiment]['Description']).lower()
    words = text.split()
    common_words = Counter(words).most_common(n)
    return pd.DataFrame(common_words, columns=['Word', 'Frequency'])

sentiments = df['Sentiment'].unique()


# 5. Word Clouds by Sentiment
for sentiment in sentiments:
    text = " ".join(df[df['Sentiment'] == sentiment]['Description'].astype(str))
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
    plt.figure(figsize=(10,5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(f"WordCloud for '{sentiment}' Sentiment")
    plt.show()

# 6. Top Nouns per Sentiment
def get_top_nouns(sentiment, n=10):
    text = " ".join(df[df['Sentiment'] == sentiment]['Description']).lower()
    tokens = word_tokenize(text)
    filtered = [word for word in tokens if word.isalpha() and word not in stop_words]
    pos_tags = pos_tag(filtered)
    nouns = [word for word, tag in pos_tags if tag.startswith('NN')]  # NN, NNS, NNP, NNPS
    common_nouns = Counter(nouns).most_common(n)
    return pd.DataFrame(common_nouns, columns=['Noun', 'Frequency'])

# Display top nouns
for sentiment in sentiments:
    print(f"\nTop Nouns in '{sentiment}' Sentiment Descriptions:")
    print(get_top_nouns(sentiment))

# Plot top nouns for each sentiment
for sentiment in sentiments:
    noun_df = get_top_nouns(sentiment)
    plt.figure(figsize=(8, 4))
    sns.barplot(data=noun_df, x='Noun', y='Frequency', palette='mako')
    plt.title(f"Top Nouns in '{sentiment}' Sentiment Descriptions")
    plt.xlabel("Noun")
    plt.ylabel("Frequency")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

```
